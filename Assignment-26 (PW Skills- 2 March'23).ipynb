2) What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

- Scatter plot is nothing but the virtual representation of data through dots or any other symbols for each and every data points of a dataset.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(3,3))
plt.scatter(x,y,c='y',alpha=0.5)
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.title("Relationship between x and y")
plt.grid()
```

3) Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

subplot() function of matplotlib library is used to create a figure and a set of subplots.

```python
import numpy as np
```

```python
fig, ax = plt.subplots(2,2)

ax[0][0].plot(np.array([0, 1, 2, 3, 4, 5]), np.array([0, 100, 200, 300, 400, 500]))
ax[0][0].set_title("Line 1")
ax[0][1].plot(np.array([0, 1, 2, 3, 4, 5]), np.array([50, 20, 40, 20, 60, 70]))
ax[0][1].set_title("Line 2")
ax[1][0].plot(np.array([0, 1, 2, 3, 4, 5]), np.array([10, 20, 30, 40, 50, 60]))
ax[1][0].set_title("Line 3")
ax[1][1].plot(np.array([0, 1, 2, 3, 4, 5]), np.array([200, 350, 250, 550, 450, 150]))
ax[1][1].set_title("Line 4")
```

4) What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

Bar plot is a graph in which data is represented in the form of pillors or bars.
- It is used when we want to know that how many data points are available for different categories.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(3,3))
plt.bar(company, profit, color='r')
plt.xlabel("Company names")
plt.ylabel("Profit for each company")
plt.title("Profit report")
plt.grid()
```

```python
plt.figure(figsize=(3,3))
plt.barh(company, profit, color='r')
plt.xlabel("Company names")
plt.ylabel("Profit for each company")
plt.title("Profit report")
plt.grid()
```

5) What is a box plot? Why is it used? Using the following data plot a box plot.

In a box plot, a box is created from first quartile to third quartile, a vertical line is also there which goes through the box at the median.
X axis denotes the data to be plotted and Y axis denotes the frequency distribution.

A Box plot is basically created to display summary of the dataset having properties like minimum, first quartile, third quartile, median and maximum etc.

```python
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200) 
```

```python
fig = plt.figure(figsize=(6,6))
plt.boxplot(box1,notch=True,patch_artist=True)
plt.show()
```

```python
plt.boxplot(box2,notch=True,patch_artist=True)
plt.show()
```
