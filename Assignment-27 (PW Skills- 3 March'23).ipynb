1) Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

- Scatterplot: It is used when we want to visualize a dataset having large number of data points. It represent each and every data points in a form of dot/any other pattern.
- Histogram: It is useful for representing frequency count for various categories.
- relation_plot: It is useful when we want to do analysis or visualization based on one or more conditions.
- cat_plot: It is used when we want to analysis or visualizations between two attributes. One is having categorical data whereas other one is having numerical data.
- joinplot: It gives us a scatterplot as well as seperate histograms for both of the attributes.

2) Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
```

```python
data_set = sns.load_dataset('fmri')
d1 = data_set[(data_set['event'] == 'stim') & (data_set['region'] == 'parietal')]
d2 = data_set[(data_set['event'] == 'cue') & (data_set['region'] == 'parietal')]
d3 = data_set[(data_set['event'] == 'cue') & (data_set['region'] == 'frontal')]
d4 = data_set[(data_set['event'] == 'stim') & (data_set['region'] == 'frontal')] 
```

```python
# This is combine graph
sns.lineplot(x=data_set.timepoint, y=data_set.subject, hue=data_set.event, style=data_set.region)
```

```python
sns.lineplot(x=d1.timepoint, y=d1.subject)
```

```python
sns.lineplot(x=d2.timepoint, y=d2.subject)
```

```python
sns.lineplot(x=d3.timepoint, y=d3.subject)
```

```python
sns.lineplot(x=d4.timepoint, y=d4.subject)
```

3) Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
data_set = sns.load_dataset('titanic')
```

```python
sns.boxplot(x='pclass', y='age', data=data_set)
```

```python
sns.boxplot(x='pclass', y='fare', data=data_set)
```
