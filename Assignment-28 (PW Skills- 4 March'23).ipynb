1) Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.graph_objects as go
import plotly as pt

data_set = sns.load_dataset('titanic')
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=data_set.age, y=data_set.fare, mode='markers'))
```

2) Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
tips = pt.data.tips()
```

```python
import plotly.express as px

fig = px.box(tips, x='total_bill', y='tip')
fig.show()
```

3) Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
tips = pt.data.tips()
```

```python
px.histogram(x='sex', y='total_bill', data_frame=tips, color='day', pattern_shape='smoker')
```

4) Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.

```python
iris = pt.data.iris()
lst = ['sepal_length','sepal_width','petal_length','petal_width']
```

```python
px.scatter_matrix(data_frame=iris, color='species', dimensions=lst)
```

5) What is Distplot? Using Plotly express, plot a distplot.
- The distplots in plotly python combines various statistical distribution such as histograms, violin plot, box plot etc in the same plot.

```python
px.histogram(tips, x='total_bill', y='tip', color='sex', marginal='box')
```
