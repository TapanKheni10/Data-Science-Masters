**1) Explain the difference between simple linear regression and multiple linear regression. Provide an example of each.**

**Simple Linear Regression**

Simple Linear Regression is a statistical method used to model the relationship between a single independent variable and a dependent variable. The relationship is represented by a linear equation in the form: **y=mx+b**

The goal of simple linear regression is to find the best-fitting line that minimizes the sum of squared differences between the observed and predicted values.

Let's consider a scenario where we want to predict a student's final exam score (y) based on the number of hours they spent studying (x). 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

hours_studied = np.array([2, 4, 6, 8, 10, 12]).reshape(-1, 1)
exam_scores = np.array([60, 70, 80, 85, 90, 95])

model = LinearRegression()

model.fit(hours_studied, exam_scores)

## Predict the exam score for unseen data of hours spent on study
new_hours = np.array([5, 7]).reshape(-1, 1)
predicted_score = model.predict(new_hours)

plt.scatter(hours_studied, exam_scores, label="Actual Data")
plt.plot(hours_studied, model.predict(hours_studied), color="red", label="Regression Line")
plt.scatter(new_hours, predicted_score, color="green", label="Predicted Data")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.legend()
plt.grid()
plt.show()
```

**Multiple Linear Regression**

Multiple Linear Regression extends simple linear regression to multiple independent variables. The relationship is represented by the equation: **y=b<sub>0</sub> + b<sub>1</sub>x<sub>1</sub> + b<sub>2</sub>x<sub>2</sub> + .... + b<sub>n</sub>x<sub>n</sub>**

Multiple linear regression aims to find the best-fitting hyperplane in a multidimensional space that minimizes the sum of squared differences between observed and predicted values.

Consider a scenario where we want to predict the price of a house (y) based on multiple features such as the number of bedrooms, square footage, and the neighborhood's safety rating.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

data = {
    "bedrooms":[2,3,4,3,4],
    "square_footage":[1200,1500,1800,1350,1600],
    "safety_rating":[8,7,9,8,9],
    "price":[200000,250000,300000,240000,280000]
}

df = pd.DataFrame(data)

X = df.drop(["price"], axis=1)
y = df["price"]

model_2 = LinearRegression()

model_2.fit(X, y)

## Coefficients and intercepts
print("Coeffieicents:", model_2.coef_)
print("Intercept:", model_2.intercept_)
```

**2) Discuss the assumptions of linear regression. How can you check whether these assumptions hold in a given dataset?**

Linear regression makes several assumptions about the underlying data.Violations of these assumptions can impact the reliability of the regression analysis.

Here are the key assumptions of linear regression:

1) Linearity:
- Assumption: The relationship between the independent and dependent variables is linear.
- Check: Use scatter plot to visualize the relationship between each independent variable and the dependent variable. (Roughly straight line)
2) Independence of Errors:
- Assumption: The errors (residuals) should be independent of each other. The error of one observation should not predict the error of another.
- Check: Examine residual plots for patterns or trends. There should be no systematic relationship between the residuals and any independent variable.
3) Homoscedasticity (Constant Variance of Errors):
- Assumption: The variance of the errors should be constant across all the level of the independent variables.
- Check: Plot residuals against predicted values. The spread of residuals should be roughly constant across the range of predicted values.
4) No Perfect Multicollinearity:
- Assumption: There should not be a perfect linear relationship between independent variables (multicollinearity).
- Check: Calculate the variance inflation factor (VIF) for each independent variable. High VIF values indicate potential multicollinearity.

**3) How do you interpret the slope and intercept in a linear regression model? Provide an example using a real-world scenario.**

Interpretetion of Slope(β):

The slope(β) represents the change in the dependent variable (Y) for a one unit change in the corresponding independent variable while holding other variables constant.

Interpretetion of Intercept(β<sub>0</sub>):

The intercept(β<sub>0</sub>) represents the expected value of the dependent variable(Y) when all independent variables are set to zero.

Example with a Real-World Scenario:

Let's consider a real-world scenario of predicting house prices based on the square footage of the house.

- Dependent Variable(Y): House Price
- Independent Variable(X): Square Footage

Assuming the linear regression equation is: **House Price = β<sub>0</sub> + β<sub>1</sub> * Square Footage + ε**

- If the square footage(X) is zero, the intercept(β<sub>0</sub>) represents the estimated house price. However, interpreting this intercept might not be meaningful in this context because a house with zero square footage doesn't exist.
- If β<sub>1</sub> for example, 100, it means that for every additional square foot in the house, the estimated house price increases by $100, assuming all other factors are constant.

**4) Explain the concept of gradient descent. How is it used in machine learning?**

Gradient descent is an optimization algorithm used to minimize the cost function or error of a model by adjusting its parameters. It is a first-order iterative optimization algorithm that finds the minimum of a function.

In the context of machine learning, gradient descent is commonly used to update the parameters (weights and biases) of a model during the training process. The goal is to find the optimal set of parameters that minimizes the cost function, which measures the difference between the model's predictions and the actual values in the training data.

**5) Describe the multiple linear regression model. How does it differ from simple linear regression?**

Multiple linear regression is an extension of simple linear regression, allowing for the modeling of relationships between a dependent variable and multiple independent variables.In a multiple linear regression model, the relationship is expressed as a linear equation: **y=β<sub>0</sub> + β<sub>1</sub>x<sub>1</sub> + β<sub>2</sub>x<sub>2</sub> + .... + β<sub>n</sub>x<sub>n</sub>**

Differences from Simple Linear Regression:
1) Number of Independent Variables:
- Simple Linear Regression: Involves only one independent variable.
- Multiple Linear Regression: Involves two or more independent variables.
2) Interpretation of Coefficients:
- Simple Linear Regression: The slope(β<sub>1</sub>) represents the change in Y for a one-unit change in X.
- Multiple Linear Regression: Each β<sub>i</sub>  represents the change in Y for a one-unit change in the corresponding X<sub>i</sub>, while holding other variables constant.

**6) Explain the concept of multicollinearity in multiple linear regression. How can you detect and address this issue?**

Multicollinearity occurs when two or more independent variables in a multiple linear regression model are highly v=correlated, making it difficult to distinguish the individual effects of each variable on the dependent variables. This can lead to unstable and unreliable estimates of the regression coefficients.

Detection of Multicollinearity:
1) Correlation Matrix:
- Examine the correlation matrix to identify pairs of variables with high correlation coefficients.
2) Variance Inflation Factor (VIF):
- VIF measures how much the variance of a regression coefficient is increased due to multicollinearity. A high VIF (typically > 10) indicates a problematic level of multicollinearity.

Addressing the Multicolllinearity:
1) Remove One of the Correlated Variables:
- If two variables are highly correlated, consider removing one of them from the model.
2) Combine Variables:
- Instead of using individual variables, consider creating composite variables or interaction terms.
3) Feature Selection:
- Use feature selection techniques to identify and keep only the most relevant variables.
4) Principal Component Analysis (PCA):
- Apply PCA to transform the original variables into a set of uncorrelated variables (principal components).
5) Ridge Regression or Lasso Regression:
- Regularization techniques like Ridge (L2 regularization) and Lasso (L1 regularization) can help mitigate multicollinearity by penalizing large coefficients.

**7) Describe the polynomial regression model. How is it different from linear regression?**

Polynomial regression is a type of regression analysis where the relationship between the independent variable X and the dependent variable Y is modeled as an n-th degree polynomial.The polynomial equation is given by: **y=β<sub>0</sub> + β<sub>1</sub>x + β<sub>2</sub>x<sup>2</sup> + .... + β<sub>n</sub>x<sup>n</sup>**

In essence, polynomial regression allows for a curved relationship between the independent and dependent variables, capturing more complex patterns in the data than simple linear regression.

Differences from Linear Regression:
1) Nature of Relationship:
- Linear Regression: Assumes a linear relationship between X and Y, forming a straight line.
- Polynomial Regression: Allows for a more flexible, curved relationship, capturing non-linear patterns.
2) Complexity:
- Linear Regression: Simpler model with a straight-line fit.
- Polynomial Regression: More complex model that can fit curves and capture intricate patterns.
3) Degree of Polynomial:
- Linear Regression: Degree 1 polynomial.
- Polynomial Regression: Can be of any degree n, allowing for a higher degree of flexibility.
4) Overfitting Risk:
- Linear Regression: Generally less prone to overfitting.
- Polynomial Regression: Higher degrees may lead to overfitting, capturing noise in the data.
